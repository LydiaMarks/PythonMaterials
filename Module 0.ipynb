## Comments

Comments are used to add documentation to your program. It's very important to leave yourself, and anyone else that has to use your script, notes on how to use the program and, in complex cases, how you did things.

### In-line comments
Use in-line comments to add documentation to explain what the code is doing. 

We're going to go over these more later, but a few common reasons to use in-line comments are: 
- explain what a variable is
- describe what a function does
- detail mathematical operations

```python
# this is an example of a comment
# we can document information about our program using comments

# this variable (x) is set to 5
x = 5

# a function is a code block that takes in some number of arguments and returns a value
# this function takes two numeric variables and adds them together
def add_vars(var1, var2):
    new_var = var1 + var2
    return(new_var)

# this function multiply input_var by 2
def times_two(input_var):
    new_var = input_var * 2
    return(new_var)
```

## Troubleshooting: When Stuff Goes Wrong
## Use Print Statements

```python
def long_function(var1, var2):
    new_var = var1 + var2
    return(new_var)
```

This is a function. It is a set of instructions that takes in arguments (the variables inside parenthesis) and does something with them. In this case the two variables are summed and the total is returned. When a value is returned it is not printed to the screen, instead the returned value is made available to the function that called this function. But something isn't right, so we need to make sure that the values coming in, and the value going out, are what we expect.

```python
def long_function(var1, var2):
    new_var = var1 + var2
    # return(new_var)
    print(var1)
    print(var2)
    print(new_var)
```

Note that the return statement is commented out. Instead we are printing both the incoming variables and the output variable. Since there is no return value we can also expect that the remainder of the program will stop running (this may give other errors, which is expected).

Another way of doing this is with an "f-string" (which is Python shorthand for "formatted string", and some languages call these "string literals"). If I had the above code and needed to see what the variables are it would look like this:

```python
def long_function(var1, var2):
    new_var = var1 + var2
    # return(new_var)
    print(f"************************ var1 = {var1} and var2 = {var2} and new_var = {new_var}")
```

Including your initials or a string of one repeated character makes your print statement easier to find when your output is in a terminal window. Without prepending f to the string you can print variables after the closing quotation marks, separating all the elements with commas:

```python
    print("************************ var1, var2, and new_var = ", var1, var2, new_var)
```

### Peers / Rubber Ducking
"A problem well-stated is a problem half solved" according to Charles Kettering (though the original quote seems lost to history). For our purposes, when asking a coworker to help with a specific problem the usual advice is to

* State what you're trying to accomplish
* State what you've tried
* State what the code is doing (or not doing)

Related advice is to say the answers to those questions aloud to something inanimate, traditionally a rubber duck. Often, gathering your thoughts sufficiently to describe the problem to someone else is enough for your brain to realize the error.
